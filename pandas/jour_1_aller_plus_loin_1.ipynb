# Partie 2 

Executer la cellule suivante pour créer le dataframe de reference

````python
import pandas as pd
import numpy as np

noms = ['Julien', 'Bastien', 'Fanny', 'David', 'Eva', 'Solene', 'Karim', 'Anna', 'Mamadou', 'Julia']
df = pd.DataFrame({
    'nom': np.random.choice(noms, size=300),
    'age': np.random.randint(20, 64, size=300),
    'niveau': np.random.choice(['Junior', 'Senior'], size=300),
    'salaire': np.array([np.nan]*100 + list(np.random.randint(25000, 60000, size=200))),
    'taille': np.random.normal(1.72, 0.1, size=300),
    'adresse': np.array([np.nan]*50 + list(np.random.choice(['Villeurbanne', 'Lyon', 'Caluire', 'Venissieux'], size=250)))
})

df
````

### Partie 2 questions

> 1/ Créer deux nouveaux dataframes en filtrant par le niveau: un dataframe niveau_senior 'filtré par Senior' et un dataframe niveau_junior 'filtré par Junior'

> 2/ Créer une nouvelle colonne "niveau salaire" avec la condition suivante: si le salaire est supérieur à 35000, alors ecrire "Très bon salaire", sinon ecrire "Salaire moyen"

> 3/ Calculer la taille minimale et maximale présent dans le dataframe pour les personnes travaillant sur Lyon. Il faudra avant cela filtré le dataframe pour obtenir uniquement les personnes travaillant sur Lyon.

> 4/ Calculer les metrics moyennes (salaire, age, taille) par ville. Il faudra pour cela effectuer un groupby par ville, en ajoutant le calcule de la moyenne.

> 5/ Effectuer une boucle sur la colonne "niveau". Afficher, "Expertise technique forte" si la valeur est "Senior" et "Expertise technique faible" si la valeur est "Junior". Afficher egalement le numero de tour de chaque boucle.

Exemple de sortie question 5:

"Tour 0: Expertise technique forte"

"Tour 1: Expertise technique faible"

````python
### code question 1
filtre_par_senior = df[df["niveau"] == "Senior"]
filtre_par_junior = df[df["niveau"] == "Junior"]

print(filtre_par_senior)
print(filtre_par_junior)
````

````python
### code question 2
df["niveau_salaire"] = np.where(df["salaire"] > 35000, "Très bon salaire", "Salaire moyen")
df
````

````python
### code question 3
df_lyon = df[df["adresse"] == "Lyon"]
min = df_lyon["taille"].min()
max = df_lyon["taille"].max()

print('min:', min, '; max', max)
````

````python
### code question 4
df.groupby("adresse").agg({"salaire": "mean", "taille": "mean", "age": "mean"})
````

````python
### code question 5
for index, niveau in enumerate(df["niveau"]):
    if niveau == "Senior":
        expertise_str = "Expertise technique forte"
    else:
        expertise_str = "Expertise technique faible"

    print(f"Tour {index}: {expertise_str}")
````

*Par curiosité, j'ai été demandé à Claude s'il existait des moyens plus efficaces de faire cette boucle en utilisant `pandas`:*

Oui, il y a plusieurs façons plus efficaces de faire cela avec pandas. Voici les meilleures approches :

1. Utiliser `enumerate` directement sur la Series avec une compréhension de liste :
```python
[f"Tour {i}: {'Expertise technique forte' if niv == 'Senior' else 'Expertise technique faible'}"
 for i, niv in enumerate(df["niveau"])]
```

2. Encore plus "pandas", utiliser `np.where` avec l'index :
```python
expertise = np.where(df["niveau"] == "Senior",
                    "Expertise technique forte",
                    "Expertise technique faible")
for i, exp in enumerate(expertise):
    print(f"Tour {i}: {exp}")
```

3. La plus élégante et efficace, en chaînant les opérations pandas :
```python
(df["niveau"]
 .map({"Senior": "Expertise technique forte",
       "Junior": "Expertise technique faible"})
 .reset_index()
 .apply(lambda x: f"Tour {x.name}: {x[1]}", axis=1)
 .to_list())
```

La troisième approche est la plus "pandas" et évite une boucle explicite, ce qui la rend plus performante sur de grands jeux de données. Elle utilise les fonctionnalités natives de pandas comme `map` pour la transformation et `apply` pour le formatage.

Quelle que soit l'approche choisie, il suffit d'ajouter un `print` pour chaque élément de la liste résultante pour obtenir exactement le format de sortie demandé.

````python
[f"Tour {i}: {'Expertise technique forte' if niv == 'Senior' else 'Expertise technique faible'}"
 for i, niv in enumerate(df["niveau"])]
````

````python
expertise = np.where(df["niveau"] == "Senior",
                    "Expertise technique forte",
                    "Expertise technique faible")
for i, exp in enumerate(expertise):
    print(f"Tour {i}: {exp}")
````
