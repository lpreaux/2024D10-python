### Featuring

#### Filtrer le jeu de donnée pour obtenir uniquement les departements avec une population superieur à 10% de la moyenne de l'ensemble des categories (H et F tout age)

```python
# votre code ici
insee['total_pop'] = insee.filter(regex='^[HF]_').sum(axis=1)
insee
```

```python
# votre code ici
seuil = insee['total_pop'].mean() * 1.10    # Pas sûr de la consigne "les departements avec une population superieur à 10% de la moyenne de l'ensemble des categories"
                                            # ça pourrait vouloir dire `> moyen_national * 0.10` mais tous les départements entre dans ce filtre
                                            # donc j'ai fait tous départements dépassant de plus de 10% de la moyenne ce qui permet d'avoir réellement un filtre.
mask = insee.total_pop > seuil
filtered_insee = insee[mask]
filtered_insee
```

#### Filtrer le jeu de donnée pour obtenir 2 nouveau df: un pour les hommes et un pour les femmes
Selectionner 3 departements de votre choix dans chaque dataframe puis merger ces dataframe a partir de ces 3 departements

Effectuer ensuite le ratio H/F sur les personne ayant jusqu'39 ans

```python
# votre code ici
insee_homme = insee.filter(regex='^H_')
insee_homme
```

```python
# votre code ici
insee_femme = insee.filter(regex='^F_')
insee_femme
```

```python
# votre code ici
insee_homme_3dep = insee_homme.loc[['Loire-Atlantique', 'Hérault', 'Paris']]
insee_femme_3dep = insee_femme.loc[['Loire-Atlantique', 'Hérault', 'Paris']]
print(insee_homme_3dep)
print(insee_femme_3dep)
```

```python
# votre code ici
insee_hf_3dep = insee_homme_3dep.join(insee_femme_3dep)
insee_hf_3dep
```

```python
# votre code ici
totaux_hf_3dep = insee_hf_3dep.filter(regex='^H_').values + insee_hf_3dep.filter(regex='^F_').values
insee_hf_totaux = pd.DataFrame(totaux_hf_3dep,
                        columns=[f'TOT_{col}' for col in insee_homme_3dep.columns.str.replace('H_', '')],
                        index=insee_hf_3dep.index)
insee_hf_totaux
```

```python
ratios = insee_hf_3dep.filter(regex='^H_').values / insee_hf_3dep.filter(regex='^F_').values
insee_hf_ratios = pd.DataFrame(ratios,
                        columns=[f'RATIO_{col}' for col in insee_homme_3dep.columns.str.replace('H_', '')],
                        index=insee_hf_3dep.index)
insee_hf_ratios
```

```python
insee_concat = pd.concat([insee_hf_3dep, insee_hf_totaux, insee_hf_ratios], axis=1)
insee_concat
```

```python
# Créer des listes pour chaque catégorie d'âge
categories = ['0_19_ans', '20_39_ans']
colonnes_ordonnees = []

for cat in categories:
    colonnes_ordonnees.extend([
        f'H_{cat}',
        f'F_{cat}',
        f'TOT_{cat}',
        f'RATIO_{cat}'
    ])

# Réorganiser les colonnes
insee_ratio_moins_39 = insee_concat.reindex(columns=colonnes_ordonnees)
insee_ratio_moins_39
```

#### Visualisation
Faire un graphqiue en bar (histogramme) representant le nombre de personne 20_39 ans par region, a partir du dataframe créer precedemment

Utiliser matplotlib pour cela 

https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py

installation de matplotlib: !pip install matplotlib

```python
## vos codes ici
from matplotlib import pyplot as plt, ticker
print(plt.style.available)
plt.style.use('_mpl-gallery')

# Trier les données avant de créer le graphique
region_insee['total'] = region_insee['H_20_39_ans'] + region_insee['F_20_39_ans']
region_insee_sorted = region_insee.sort_values('total', ascending=False)
```

```python
## vos codes ici
fig, ax = plt.subplots(figsize=(15, 8))

# Position des barres
x = np.arange(len(region_insee_sorted.index))
width = 0.35

# Création des barres
ax.bar(x - width/2, region_insee_sorted['H_20_39_ans']/1e6, width, label='Hommes', color='skyblue', edgecolor='navy')
ax.bar(x + width/2, region_insee_sorted['F_20_39_ans']/1e6, width, label='Femmes', color='red', edgecolor='red')

# Personnalisation
ax.set_xticks(x)
ax.set_xticklabels(region_insee_sorted.index, rotation=45, ha='right')
ax.yaxis.set_major_locator(plt.MultipleLocator(0.25))
ax.legend(framealpha=1)

# Titres et labels
plt.title('Population 20-39 ans par région et par sexe', fontsize=14)
plt.xlabel('Régions', labelpad=10)
plt.ylabel('Population (millions)')


# Ajout d'une grille pour faciliter la lecture
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Ajustement automatique de la mise en page
plt.tight_layout()

plt.show()
```
